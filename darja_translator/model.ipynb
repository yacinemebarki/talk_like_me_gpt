```python
import numpy as np
import pandas as pd
```

```python
path="/home/yacine10/talk_like_me_gpt/data/train-00000-of-00001.csv"
data=pd.read_csv(path)
data
```

```python
arabizi=data[data["includes_arabizi"]==True]
print("the number of arabizi sentence=",len(arabizi))
```

```python
data=data[data["includes_arabizi"]==True]
data.reset_index()
```

```python
data=data[["darija","english"]]
data=data.reset_index()
```

```python
data=data.drop(columns=["index"])
data
```

```python
print(len(np.unique(data["darija"])))
print(len(np.unique(data["english"])))
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras import layers,models
```

```python
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
print(dir(Tokenizer))
```

```python
x=data["darija"]
y=data["english"]
```

```python
tk_darja=Tokenizer()
tk_darja.fit_on_texts(x)
tk_english=Tokenizer()
tk_english.fit_on_texts(y)
```

```python
x_sequence=tk_darja.texts_to_sequences(x)
y_sequence=tk_english.texts_to_sequences(y)
```

```python
print(dir(pad_sequences))
```

```python
x_sequence=pad_sequences(x_sequence,maxlen=32)
y_sequence=pad_sequences(y_sequence,maxlen=32)
```

```python
x_sequence=np.array(x_sequence)
y_sequence=np.array(y_sequence)
```

```python

x_train,x_test,y_train,y_test=train_test_split(x_sequence,y_sequence,test_size=0.2,random_state=42)
```

```python
print(dir(layers))
```

```python
print(dir(models))
```

```python
print(dir(layers.RNN))
```

```python
model=models.Sequential()
model.add(layers.Embedding(input_dim=len(tk_darja.word_index)+1,output_dim=64,input_length=32))
model.add(layers.SimpleRNN(64,return_sequences=True))
model.add(layers.TimeDistributed(layers.Dense(len(tk_english.word_index)+1, activation="softmax")))
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy")
```
