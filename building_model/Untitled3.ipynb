```python
import tensorflow as tf
from tensorflow.keras import models,layers
from tensorflow.keras.layers import MultiHeadAttention,LayerNormalization
class decoder:

    def __init__(self,num_heads,key_dim):
        model=models.Sequential()
        model.add(layers.Dense(256,activation="relu"))
        model.add(layers.Dense(64))
        self.fnn=model
        self.msa=MultiHeadAttention(num_heads,key_dim)
        self.cross_att = MultiHeadAttention(num_heads, key_dim)
        self.ln1 = LayerNormalization()
        self.ln2 = LayerNormalization()


    def forward(self,x,enc_out):
        out1=self.msa(query=x,value=x,key=x,use_causal_mask=True)

        x1=self.ln1(x+out1)
        cross = self.cross_att(query=x1, key=enc_out, value=enc_out)

        out2=self.fnn(cross)
        x2=self.ln2(cross+out2)
        return x2
```

```python
class encoder:

  def __init__(self,num_heads,key_dim):
    model=models.Sequential()
    model.add(layers.Dense(256,activation="relu"))
    model.add(layers.Dense(64))
    self.fnn=model
    self.msa=MultiHeadAttention(num_heads,key_dim)
    self.ln1 = LayerNormalization()
    self.ln2 = LayerNormalization()

  def forward(self,x):
    out1=self.msa(query=x,value=x,key=x)
    x1=self.ln1(x+out1)
    out2=self.fnn(x1)
    x2=self.ln2(x1+out2)
    return x2

```

```python
from tensorflow.keras.layers import Dense,Embedding
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import Adam




class model:

    def __init__(self,n_decoder,n_encoder,input_size,num_heads,key_dim,n_token):
        self.input_size=input_size
        self.enlayers=[]
        self.delayers=[]
        self.dense=Dense(n_token,activation="softmax")
        self.emb=Embedding(input_dim=n_token,output_dim=64,input_length=32)
        self.optimizer = tf.keras.optimizers.Adam(0.01)
        self.loss_fn = SparseCategoricalCrossentropy()
        for i in range(n_encoder):
          self.enlayers.append(encoder(num_heads,key_dim))
        for i in range(n_decoder):
          self.delayers.append(decoder(num_heads,key_dim))




    def fit(self,x,y,epoch):
        n_sample=len(x)
        for ep in range(epoch):
            for i in range(n_sample):
                a = tf.expand_dims(x[i], axis=0)
                with tf.GradientTape() as tape:
                    a=self.emb(a)
                    for l in self.enlayers:
                        a=l.forward(a)
                        enc_out=a
                    for l in self.delayers:
                        a=l.forward(a,enc_out)

                    out=self.dense(a)
                    y_true = tf.expand_dims(y[i], axis=0)
                    loss=self.loss_fn(y_true, out)

                variables=self.dense.trainable_variables
                for l in reversed(self.delayers):
                    variables += l.fnn.trainable_variables
                    variables += l.msa.trainable_variables
                    variables += l.msa.trainable_variables


                for l in reversed(self.enlayers):
                    variables += l.fnn.trainable_variables
                    variables += l.msa.trainable_variables
                variables +=self.emb.trainable_variables




                grads = tape.gradient(loss, variables)
                self.optimizer.apply_gradients(zip(grads,variables))
```
