```python
from tensorflow.keras import layers,models
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
from sklearn.model_selection import train_test_split
```

```python
df=df.reset_index()
df=df.drop(columns=["index"])
```

```python
x=df["answer"]
y=df["message"]
```

```python
tok_m=Tokenizer()
tok_a=Tokenizer()
tok_m.fit_on_texts(x)
tok_a.fit_on_texts(y)
```

```python
x_seq=tok_m.texts_to_sequences(x)
y_seq=tok_a.texts_to_sequences(y)
```

```python
x_seq=pad_sequences(x_seq,maxlen=32)
y_seq=pad_sequences(y_seq,maxlen=32)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_seq,y_seq,test_size=0.2,random_state=42)
```

```python
model=models.Sequential()
model.add(layers.Embedding(input_dim=len(tok_m.word_index)+1,output_dim=32))
model.add(layers.SimpleRNN(64,return_sequences=True))
model.add(layers.TimeDistributed(layers.Dense(len(tok_a.word_index)+1,activation="softmax")))
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy")
```

```python
model.fit(x_seq,y_seq,epochs=5,batch_size=32)
```
