```python
def seq_to_text(seq, vocab):
    return [" ".join(vocab[i] for i in s if i != 0) for s in seq]

```

```python
answer_vocab=tok_a.get_vocabulary()
for i in range(len(pred)):
  print("question is",question.iloc[i],"resopnse is",seq_to_text([pred[i]],answer_vocab))
```

```python
from nltk.translate.bleu_score import sentence_bleu

references = []
candidates = []

for i in range(len(y_test)):
    ref_seq = seq_to_text([y_test[i]], answer_vocab)[0].split()
    pred_seq = seq_to_text([pred[i]], answer_vocab)[0].split()
    references.append([ref_seq])
    candidates.append(pred_seq)
```

```python
bleu_scores = []

for ref, cand in zip(references, candidates):
    score = sentence_bleu(ref, cand)
    bleu_scores.append(score)

print("Average BLEU:", sum(bleu_scores)/len(bleu_scores))
```
